```python
cols = ['log_price','Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_preprocessed[['Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['features'] = variables.columns
```

```python
vif
```

# Linear regression model

Declaring inputs and targets

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'],axis=1)
```

# Scaling the data

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)
```

# Creating a regression

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets(y_train)',size=18)
plt.ylabel('Predictions(y_hat)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.displot(y_train - y_hat)
plt.title('Residual PDF' , size=18)
```

```python
reg.score(x_train,y_train)
```

# Finding the weights and bias

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns =['Features'])
reg_summary['coefficients'] = reg.coef_
reg_summary
```

# Testing

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test,alpha=0.2)
plt.xlabel('Targets(y_test)',size=18)
plt.ylabel('Predictions(y_hat_test)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

The regression model appears to be much better at predicting higher prices than lower prices.

# Getting the exponential of the log prices to revert to the original prices.

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Prediction'])
df_pf.head()
```

```python
y_test = y_test.reset_index(drop=True)
y_test.head()
```

```python
df_pf['Input'] = np.exp(y_test)
df_pf.head()
```

```python
df_pf['Residual'] = df_pf['Input'] - df_pf['Prediction']
df_pf
```

```python
df_pf['Difference%'] = np.absolute(df_pf['Residual'])/df_pf['Input']*100
df_pf
```

```python
df_pf.describe()
```

```python
df_pf.sort_values(by=['Difference%'])
```

```python
pd.options.display.max_rows = 999
pd.set_option('display.float_format', lambda x: '%.2f' % x)
df_pf.sort_values(by=['Difference%'])
```

The year and the mileage of the car seem to have the most effect on the price of the car. The model has high accuracy except for low price cars. The low price cars are also observed to have negative residuals meaning that the car price predictions are higher than the targets. It seems that there is a missing piece of information that could improve the accuracy of predicting the car prices.
